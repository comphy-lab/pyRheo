# Tutorial: Fitting Noisy Stress Relaxation Data with a Fractional Kelvin–Voigt Model

This tutorial walks through a complete workflow for loading stress relaxation experiments and fitting a **Fractional Kelvin–Voigt model** using **pyRheo**. In this case, the stress relaxation measurements exhibit a significant noise at short times. The model is fitted to the noisy data by minimizing the residual sum of square weighted by the standard deviation of each measuring point.

---

## 1. Import required libraries

We start by importing numerical, scientific, and plotting libraries, along with the `RelaxationModel` class from **pyRheo**.

```python
# %%
import numpy as np                      # Numerical array operations
import pandas as pd                     # DataFrame-based data handling
import matplotlib.pyplot as plt          # Plotting library
from scipy.special import gamma          # Special mathematical functions
from scipy.optimize import minimize      # Optimization routines
from pyRheo.relaxation_model import RelaxationModel  # Rheological model class
```

## 2. Load stress relaxation experiments

Each CSV file corresponds to one experimental replicate.
We assume:
- All experiments were performed under identical conditions
- The time vector is identical for all files
- Files are tab-separated, use decimal commas, and UTF-16 encoding

```python
# List of experimental CSV files (one per replicate)
files = [
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_2.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_3.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_5.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_6.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_7.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_8.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_9.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_10.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_11.csv",
    "./data/cnf_hydrogel/HC1_stressrelaxation_20C_12.csv",
]

# Read each CSV file into a Pandas DataFrame
# skiprows removes instrument metadata rows
data_list = [
    pd.read_csv(
        f,
        skiprows=(0, 1, 2, 4, 5),
        delimiter="\t",
        decimal=",",
        encoding="utf-16",
    )
    for f in files
]

# Extract the time vector from the first file
# (assumed identical across all replicates)
time = data_list[0]["Time"].values

# Stack relaxation modulus data into a 2D array:
# shape = (number of replicates, number of time points)
G_all = np.vstack([df["Relaxation Modulus"].values for df in data_list])

# Store dimensions for later reference
n_replicates, n_timepoints = G_all.shape
```

## 3. Filter the data (optional)

Raw rheological data often contain unreliable regions.
Here we apply two filters:
1. Remove very early times (instrument inertia / transient effects)
2. Remove very small modulus values (noise floor)
   
Time points where *all* replicates are invalid are discarded.


```python
# Keep only time points greater than or equal to 0.05 s
mask_time = time >= 2
time_filtered = time[mask_time]
G_filtered = G_all[:, mask_time]

# Replace modulus values below 1 Pa with NaN (mark as invalid)
G_filtered_masked = np.where(G_filtered >= 1, G_filtered, np.nan)

# Identify time points where at least one replicate is valid
valid_cols = ~np.all(np.isnan(G_filtered_masked), axis=0)

# Keep only valid time points
time_valid = time_filtered[valid_cols]
G_valid = G_filtered_masked[:, valid_cols]
```

## 4. Compute mean and standard deviation

We compute statistics across experimental replicates:
- Mean relaxation modulus (used for fitting)
- Standard deviation (used as a weighting factor)
  
Weighting by experimental uncertainty improves robustness of the fit. The user can modify the cost function to visualize this.

```python
# Compute mean relaxation modulus across replicates
G_mean = np.nanmean(G_valid, axis=0)

# Compute sample standard deviation (ddof=1 for unbiased estimate)
G_std = np.nanstd(G_valid, axis=0, ddof=1)

# Prevent zero standard deviation (would cause infinite weighting)
G_std[G_std == 0] = np.min(G_std[G_std > 0])
```

## 5. Sort data by time

Sorting ensures strictly increasing time values, which is required by many pyRheo and plotting routines.


```python
# Get indices that sort the time array
sort_idx = np.argsort(time_valid)

# Apply the same sorting to all data arrays
time_sorted = time_valid[sort_idx]
G_mean_sorted = G_mean[sort_idx]
G_std_sorted = G_std[sort_idx]
G_valid_sorted = G_valid[:, sort_idx]
```

## 6. Initialize the fractional Kelvin–Voigt model

 We configure a pyRheo relaxation model with:
 - Fractional Kelvin–Voigt formulation
 - Random initial parameter guesses
 - Powell optimization (gradient-free)
 - Weighted residuals using experimental standard deviation


```python
model = RelaxationModel(
    model="FractionalKelvinVoigtS",        # Fractional Kelvin–Voigt model
    initial_guesses="random",              # Randomized starting parameters
    minimization_algorithm="Powell",       # Robust, derivative-free optimizer
    cost_function=("RSS_custom", G_std_sorted),  # Weighted residual sum of squares, on the second argument you must give the vector of the same shape as G_mean_sorted. In SOAS, you have to give two arguments, for the storage and loss modulus
    #cost_function="RSS"
)
```

## 7. Fit the model

The model is fitted to the mean relaxation modulus as a function of time.


```python
# Perform model fitting
model.fit(time_sorted, G_mean_sorted)

# Retrieve optimized parameters
params = model.get_parameters()

# Print fitted parameters and fitting error
model.print_parameters()
model.print_error()

# Extract parameters from the fitted model
params = model.get_parameters()

G_p = params["G_p"]
V = params["V"]
alpha = params["alpha"]

# Retardation time
tau_r = (V / G_p) ** (1 / alpha)
print(f"Retardation time τ_r = {tau_r:.4f} s")
```

## 8. Predict and visualize the fitted relaxation model to the data

The model is fitted to the mean relaxation modulus as a function of time.


```python
# Predict relaxation modulus from the fitted model
# flatten() ensures a 1D array for plotting
time_predict = np.logspace(-0,2, 100)
G_fit = model.predict(time_predict).flatten()


plt.figure(figsize=(6, 3))

# Generate a colormap with as many colors as replicates
cmap = plt.cm.viridis
colors = cmap(np.linspace(0, 1, G_valid_sorted.shape[0]))

# Plot each experimental replicate with a unique color
for i, color in enumerate(colors):
    plt.loglog(
        time_sorted,
        G_valid_sorted[i],
        "o",
        color=color,
        alpha=0.6,
        label=f"Experiment {i+1}",
    )

# Plot fitted model curve (thick black dashed line for contrast)
plt.loglog(
    time_predict,
    G_fit,
    "k--",
    linewidth=2.5,
    label="Fitted Fractional KV",
)

# Axis labels and formatting
plt.xlabel("Time [s]")
plt.ylabel("Relaxation Modulus [Pa]")
plt.ylim(10, 10000)

# Place legend outside to avoid clutter
plt.legend(loc="upper right", fontsize=6, ncol=2, frameon=False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 3))

# Plot experimental uncertainty
plt.fill_between(
    time_sorted,
    G_mean_sorted - G_std_sorted,
    G_mean_sorted + G_std_sorted,
    color="gray",
    alpha=0.3,
    label="Experimental ± std",
)

# Plot mean experimental data
plt.loglog(
    time_sorted,
    G_mean_sorted,
    "o",
    color="black",
    label="Mean experimental",
)

# Plot fitted model
plt.loglog(
    time_predict,
    G_fit,
    "r--",
    linewidth=2,
    label="Fitted Fractional KV",
)

# Axis labels and formatting
plt.xlabel("Time [s]")
plt.ylabel("Relaxation Modulus [Pa]")
plt.ylim(10, 10000)
plt.legend(loc="upper right", fontsize=6, ncol=2, frameon=False)
plt.show()
```
