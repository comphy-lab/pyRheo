# Oscillatory Rheology Models Visualization

This notebook explores various viscoelastic models to understand their dynamic behavior under oscillatory loading. It uses pyRheo to compute the models given fixed parameter values.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pyRheo.oscillation_evaluator import OscillationEvaluator

# Example omega values
omega = np.logspace(-4, 4, 15)

# Parameters
model = "Maxwell"
G_s = 1e0
eta_s = 1e3

# Create a plasma colormap
cmap = cm.plasma

# Create figure and axis
plt.figure(figsize=(3.5, 3))

params = [G_s, eta_s]
evaluator = OscillationEvaluator(model=model) 
    
# Compute storage and loss moduli
G_prime, G_double_prime = evaluator.compute_model(params, omega)
    
# Select a color from the plasma colormap
color = cmap(1)  # Scale the color index
    
# Plotting storage modulus (G') on the upper plot
plt.plot(omega, G_prime, 'o-', markersize=6, color=color, label="$G^{\prime}(\omega)$")
    
# Plotting loss modulus (G'') on the lower plot
plt.plot(omega, G_double_prime, 'o-', markersize=6, fillstyle='none', color=color, label="$G^{\prime \prime}(\omega)$")

# Customize the plot
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xscale("log")
plt.yscale("log")
plt.ylabel(r'$G^{\prime}(\omega)$ [Pa]', fontsize=14)
plt.legend(fontsize=13.5, loc="lower left")
plt.grid(False)

#plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pyRheo.oscillation_evaluator import OscillationEvaluator

# Example omega values
omega = np.logspace(-3, 3, 15)

# Parameters
model = "SpringPot"
V = 1e3

# Example alpha values
alpha_values = [0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1]

# Create a plasma colormap
cmap = cm.plasma

# Create figure and axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 6), sharex=True, gridspec_kw={'hspace': 0.05})

# Loop over alpha values
for i, alpha in enumerate(alpha_values):
    params = [V, alpha]
    evaluator = OscillationEvaluator(model=model)
    
    # Compute storage and loss moduli
    G_prime, G_double_prime = evaluator.compute_model(params, omega)
    
    # Select a color from the plasma colormap
    color = cmap(i / len(alpha_values))  # Scale the color index
    
    # Plotting storage modulus (G') on the upper plot
    ax1.plot(omega, G_prime, 'o-', markersize=6, label=f'$\\alpha$={alpha}', color=color)
    
    # Plotting loss modulus (G'') on the lower plot
    ax2.plot(omega, G_double_prime, 'o-', markersize=6, fillstyle='none', color=color)

# Customize the upper plot (Storage modulus)
ax1.tick_params(axis='both', which='major', labelsize=14)  # For major ticks
ax1.tick_params(axis='both', which='minor', labelsize=10)  # For minor ticks
ax1.set_xscale("log")
ax1.set_yscale("log")
ax1.set_ylabel(r'$G^{\prime}(\omega)$ [Pa]', fontsize=14)
ax1.legend(fontsize=13.5, loc="upper left", bbox_to_anchor=(1.05, 1.0))
ax1.grid(False)

# Customize the lower plot (Loss modulus)
ax2.tick_params(axis='both', which='major', labelsize=14)
ax2.tick_params(axis='both', which='minor', labelsize=10)
ax2.set_xscale("log")
ax2.set_yscale("log")
ax2.set_xlabel(r'$\omega$ [rad s$^{-1}$]', fontsize=14)
ax2.set_ylabel(r'$G^{\prime \prime}(\omega)$ [Pa]', fontsize=14)
ax2.grid(False)

#plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pyRheo.oscillation_evaluator import OscillationEvaluator

# Example omega values
omega = np.logspace(-3, 3, 15)

# Parameters
model = "FractionalZenerS"
G_p = 1e3
G = 1e3
V = 1e3

# Example alpha and beta values
alpha_values = [0.9, 0.8, 0.7, 0.6, 0.51]  
beta_values = [0.1, 0.2, 0.3, 0.4, 0.50]

# Create a plasma colormap
cmap = cm.plasma

# Create figure and axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 6), sharex=True, gridspec_kw={'hspace': 0.05})

# Loop over alpha and beta values
for i, (alpha, beta) in enumerate(zip(alpha_values, beta_values)):
    params = [G_p, G, V, alpha, beta]
    evaluator = OscillationEvaluator(model=model) 
    
    # Compute storage and loss moduli
    G_prime, G_double_prime = evaluator.compute_model(params, omega)
    
    # Select a color from the plasma colormap
    color = cmap(i / len(alpha_values))  # Scale the color index
    
    # Plotting storage modulus (G') on the upper plot
    ax1.plot(omega, G_prime, 'o-', markersize=6, label=f'$\\alpha$={alpha}, $\\beta$={beta}', color=color)
    
    # Plotting loss modulus (G'') on the lower plot
    ax2.plot(omega, G_double_prime, 'o-', markersize=6, fillstyle='none', color=color)

# Customize the upper plot (Storage modulus)
ax1.tick_params(axis='both', which='major', labelsize=14)  # For major ticks
ax1.tick_params(axis='both', which='minor', labelsize=10)  # For minor ticks
ax1.set_xscale("log")
ax1.set_yscale("log")
ax1.set_ylabel(r'$G^{\prime}(\omega)$ [Pa]', fontsize=14)
ax1.legend(fontsize=13.5, loc="upper left", bbox_to_anchor=(1.05, 1.0))
ax1.grid(False)

# Customize the lower plot (Loss modulus)
ax2.tick_params(axis='both', which='major', labelsize=14)
ax2.tick_params(axis='both', which='minor', labelsize=10)
ax2.set_xscale("log")
ax2.set_yscale("log")
ax2.set_xlabel(r'$\omega$ [rad s$^{-1}$]', fontsize=14)
ax2.set_ylabel(r'$G^{\prime \prime}(\omega)$ [Pa]', fontsize=14)
ax2.grid(False)

#plt.tight_layout()
plt.show()
```
