# Rotation Rheology Models Visualization

This notebook explores various viscoelastic models to understand their behavior under rotation loading. It uses pyRheo to compute the models given fixed parameter values.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pyRheo.rotation_evaluator import RotationEvaluator

# Example shear rate values
gamma_dot = np.logspace(-4, 4, 15)

# Parameters
model = "HerschelBulkley"
sigma_y = 1e3
k = 1e3
n = 0.6

# Create a plasma colormap
cmap = cm.plasma

# Create figure and axis
plt.figure(figsize=(7,5))

params = [sigma_y, k, n]
evaluator = RotationEvaluator(model=model) 
    
# Compute relaxation
eta = evaluator.compute_model(params, gamma_dot)
    
# Select a color from the plasma colormap
color = cmap(1)  # Scale the color index
    
# Plotting storage modulus (G') on the upper plot
plt.plot(gamma_dot, eta, 'o-', markersize=6, color=color)

# Customize the plot
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xscale("log")
plt.yscale("log")
plt.xlabel('$\dot{\gamma}$ [s$^{-1}$]', fontsize=14)
plt.ylabel(r'$\eta(\dot{\gamma})$ [Pa s]', fontsize=14)
plt.grid(False)

#plt.tight_layout()
plt.show()
```
