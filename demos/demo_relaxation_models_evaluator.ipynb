# Relaxation Rheology Models Visualization

This notebook explores various viscoelastic models to understand their behavior under relaxation loading. It uses pyRheo to compute the models given fixed parameter values.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pyRheo.relaxation_evaluator import RelaxationEvaluator

# Example time values
time = np.logspace(-4, 4, 15)

# Parameters
model = "Maxwell"
G_s = 1e0
eta_s = 1e3

# Create a plasma colormap
cmap = cm.plasma

# Create figure and axis
plt.figure(figsize=(7,5))

params = [G_s, eta_s]
evaluator = RelaxationEvaluator(model=model) 
    
# Compute relaxation
G_relax = evaluator.compute_model(params, time)
    
# Select a color from the plasma colormap
color = cmap(1)  # Scale the color index
    
# Plotting storage modulus (G') on the upper plot
plt.plot(time, G_relax, 'o-', markersize=6, color=color, label="$G(t)$")
    
# Customize the plot
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xscale("log")
plt.yscale("log")
plt.ylabel(r'$G(t)$ [Pa]', fontsize=14)
plt.legend(fontsize=13.5, loc="lower left")
plt.grid(False)

#plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from pyRheo.relaxation_evaluator import RelaxationEvaluator

# Example time values
time = np.logspace(-3, 3, 15)

# Parameters
model = "SpringPot"
V = 1e3

# Example alpha values
alpha_values = [0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1]

# Create a plasma colormap
cmap = cm.plasma

# Create a figure
plt.figure(figsize=(7, 5))

# Loop over alpha values
for i, alpha in enumerate(alpha_values):
    params = [V, alpha]
    evaluator = RelaxationEvaluator(model=model) 
    
    # Compute relaxation
    G_relax = evaluator.compute_model(params, time)

    # Select a color from the plasma colormap
    color = cmap(i / len(alpha_values))  # Scale the color index

    # Plotting relaxation modulus vs. time
    plt.plot(time, G_relax, 'o-', markersize=6, label=f'$\\alpha$={alpha}', color=color)

# Customize the plot
plt.tick_params(axis='both', which='major', labelsize=14)  # For major ticks
plt.tick_params(axis='both', which='minor', labelsize=10)  # For minor ticks
plt.xscale("log")
plt.yscale("log")
plt.xlabel('$t$ [s]', fontsize=14)
plt.ylabel('$G(t)$ [Pa]', fontsize=14)
plt.legend(fontsize=13.5, loc="upper right", bbox_to_anchor=(1.3, 1.0))
plt.grid(False)

# Show the plot
plt.show()
```
