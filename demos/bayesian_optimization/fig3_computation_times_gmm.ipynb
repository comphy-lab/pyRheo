```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
import numpy as np

plt.style.use(['science', 'nature', 'retro'])

# Load the Excel file
file_path = 'computation_times.xlsx'
df = pd.read_excel(file_path)

# List of methods
methods = ['BO', 'L-BFGS-B', 'NM', 'Powell']

# Define colors
colors = {
    'BO': '#60568A',        # dark purple
    'L-BFGS-B': '#F29E4C',  # greenish
    'NM': '#89BFA9',        # orange
    'Powell': '#E15759'     # red
}

# ----------------- Plot Times -----------------
plt.figure(figsize=(3, 2.75))
for method in methods:
    plt.scatter(df['Parameters'], df[f'{method} Time'], 
                marker='o', color=colors[method], label=method, s=36, alpha=0.7)
plt.xlabel('$p$', fontsize=14)
plt.ylabel('$t$ [s]', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=12, loc='upper left', bbox_to_anchor=(-0.05, 1), handletextpad=0.3, labelspacing=0.2)
plt.tight_layout()
plt.savefig('computation_time.pdf', dpi=1200)
plt.show()

# ----------------- Plot Iterations -----------------
plt.figure(figsize=(3, 2.75))
for method in methods:
    plt.scatter(df['Parameters'], np.log10(df[f'{method} Iterations']), 
                marker='o', color=colors[method], label=method, s=36, alpha=0.7)
plt.xlabel('$p$', fontsize=14)
plt.ylabel('$\log{n}$', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig('iterations_log10.pdf', dpi=1200)
plt.show()

# ----------------- Plot Errors -----------------
plt.figure(figsize=(3, 2.75))
for method in methods:
    plt.scatter(df['Parameters'], np.log10(df[f'{method} Error']), 
                marker='o', color=colors[method], label=method, s=36, alpha=0.7)
plt.xlabel('$p$', fontsize=14)
plt.ylabel('$\\log$ RSS$_{w_{i}=y_{i}}$', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig('error_log10.pdf', dpi=1200)
plt.show()
```
