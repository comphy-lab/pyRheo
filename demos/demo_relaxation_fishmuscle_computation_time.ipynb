# Rheological analysis of fish muscle Relaxation behavior

This notebook demonstrates the use of the `pyRheo` library to fit the Relaxation behavior of a fish muscle. The experimental data includes angular frequency (`time`) and the Relaxation Modulus (`G_relax`), which are analyzed using a rheological model. The data is from Jesús Puente-Córdova's archive.

We import the following libraries:
- `pandas`: For handling the experimental data in tabular format.
- `numpy`: For numerical operations and transformations.
- `matplotlib.pyplot`: For creating visualizations of the data and model predictions.
- `pyRheo.creep_model.RelaxationModel`: To perform rheological model fitting and analysis of the Relaxation Modulus (`G_relax`) data.



```python
# Import required libraries
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical operations
import matplotlib.pyplot as plt  # For plotting
from pyRheo.relaxation_model import RelaxationModel  # For rheological modeling
```

# Load and inspect rheological data
This cell loads experimental data from a CSV file containing:
- Angular frequency (`time`) in [rad/s].
- Relaxation Modulus (`G_relax`) in [Pa].

The data is loaded using `pandas` and the first few rows are displayed for verification.


```python
# Load rheological data from a CSV file
# The file contains time (time) and relaxation (G_relax, ) data.
data = pd.read_csv('./data/fish_muscle/stressrelaxation_fishmuscle_data.csv', delimiter='\t', decimal=".")

# Extract time and relaxation as numpy arrays
time = data['Time'].values
G_relax = data['Relaxation Modulus'].values

# Display the first few rows of the data for verification
data.head()
```

# Initialize and fit the rheological model
In this cell:
1. The `RelaxationModel` class is instantiated with parameters for model fitting:
   - Automatic model selection.
   - Random initial guesses for optimization.
   - 64 guesses for the optimizer to explore.
   - The Powell algorithm for minimization.
2. The model is fitted to the experimental data (`time`, `G_relax`, and ``).
3. Fitted model parameters and error metrics are displayed.


```python
import numpy as np
import time as timer
import matplotlib.pyplot as plt

# Function to apply a vertical shift
def apply_vertical_shift(data, shift_factor=0.02):
    """Apply a vertical shift based on the mean of the dataset."""
    shift = shift_factor * np.mean(data)  # Compute shift as a fraction of mean
    return data + shift

# Initialize the rheological model
model = RelaxationModel(
    model="auto",                   
    initial_guesses="random",       
    num_initial_guesses=10,         
    minimization_algorithm="Powell",  
    mittag_leffler_type="Pade54"
)

# Extract original dataset
time = data['Time'].values
G_relax = data['Relaxation Modulus'].values

# Lists to store dataset sizes and corresponding computation times
dataset_sizes = []
computation_times = []

# Start with the original dataset
new_time = time.copy()
new_G_relax = G_relax.copy()

for i in range(1, 30):  # From 1x to 5x dataset size
    if i > 1:
        # Duplicate dataset and apply vertical shift
        shifted_G_relax = apply_vertical_shift(G_relax, shift_factor=0.02 * (i - 1))
        
        # Append the new shifted data
        new_time = np.concatenate((new_time, time))  # Repeat time values
        new_G_relax = np.concatenate((new_G_relax, shifted_G_relax))  # Add shifted data

    # Sort the dataset before fitting
    sorted_indices = np.argsort(new_time)
    sorted_time = new_time[sorted_indices]
    sorted_G_relax = new_G_relax[sorted_indices]

    # Record dataset size
    dataset_size = len(sorted_time)
    dataset_sizes.append(dataset_size)

    # Fit the model and measure time
    start_time = timer.time()
    model.fit(sorted_time, sorted_G_relax)
    end_time = timer.time()

    # Plotting the data and the model fit after each iteration
    plt.figure(figsize=(8, 5))
    
    # Plot the raw data
    plt.plot(sorted_time, sorted_G_relax, 'o', label="Data (Shifted)")

    # Plot the fitted curve
    fitted_G_relax = model.predict(sorted_time)  # Get the fitted values from the model
    plt.plot(sorted_time, fitted_G_relax, 'r-', label="Fitted Model")

    plt.xlabel('Time')
    plt.ylabel('Relaxation Modulus')
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f"Iteration {i} - Dataset Size: {dataset_size}")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Compute elapsed time and store
    elapsed_time = end_time - start_time
    computation_times.append(elapsed_time)

    print(f"Iteration {i}: Dataset size = {dataset_size}, Time taken = {elapsed_time:.4f} sec")

# Plot dataset size vs. computation time
plt.figure(figsize=(8, 5))
plt.plot(dataset_sizes, computation_times, marker='o', linestyle='-', color='b')
plt.xlabel("Dataset Size")
plt.ylabel("Computation Time (s)")
plt.title("Dataset Size vs Computation Time for Model Fitting")
plt.grid(True)
plt.show()
```

```python
# Plotting the data and the model fit after each iteration
plt.figure(figsize=(3.2, 3))
    
# Plot the raw data
plt.plot(sorted_time, sorted_G_relax, 'o', label="noisy data")

# Plot the fitted curve
fitted_G_relax = model.predict(sorted_time)  # Get the fitted values from the model
plt.plot(sorted_time, fitted_G_relax, 'r-', label="fit")

plt.xlabel('$t$ [s]', fontsize=14)
plt.ylabel('$G(t)$ [Pa]', fontsize=14)
plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.savefig("fitting_of_augmented_size_of_dataset.pdf")
plt.show()

# Plot dataset size vs. computation time
plt.figure(figsize=(3.2, 3))
plt.plot(dataset_sizes, computation_times, marker='o', ls="none", color='b')
plt.xscale('log')
plt.xlabel("dataset size", fontsize=14)
plt.ylabel("computation time [s]", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(False)
plt.tight_layout()

plt.savefig("computation_time_vs_size_of_dataset.pdf")
plt.show()
```

# Generate and plot model fit
In this cell:
1. Predictions are generated using the fitted model at logarithmically spaced times.
2. The experimental data and model predictions are visualized using a quick plot from the `pyRheo` library.


```python
# Generate predictions using the fitted model for a smooth curve
time_predict = np.logspace(np.min(np.log10(time)), np.max(np.log10(time)), 100)
G_relax_predict = model.predict(time_predict)

# Plot the experimental data and the model prediction
model.plot(time, G_relax, savefig=False)  # Quick overview plot
```
